```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
plt.plot([0, 1, 2, 3])
plt.show()
```

```python
# Make the random function consostent amd replicable.
np.random.seed(1221)

# Make the blank data frame.
df = pd.DataFrame()

# Add a column of random numbers between 0 and 1.
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2 

# When creating a data frame an indez column of counts is created, counting from 0.
# Here we do a few transforms on that indexc to create some extra columns.
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand'], color='blue')
plt.ylim([-0.1, 1.1])
plt.ylabel('Values')
plt.title('Top Secret Series')
plt.show()
```

```python
plt.plot(df['rand'], color='red')
plt.plot(df['rand_shift'], color='blue')
plt.ylim([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Top Secret Shifted')
plt.show()
```

```python
plt.scatter(x=df['rand'], y=df['rand_sq'])
plt.show()
```

```python
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='green',
    marker='x', s=10
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color='purple',
marker='x', s=10
)
plt.show()
```
