```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv('insurance.csv')
data.head()
```

```python
sns.set(style='whitegrid')

sexgraph = sns.factorplot(x='region', y='charges', data=data, kind='bar')
```

```python
bmiage = sns.lmplot(y='bmi',
                   x='age',
                   data=data,
                   fit_reg=True,
                   scatter_kws={'alpha':0.1})
bmiage.set_ylabels('BMI')
bmiage.set_xlabels('age')
plt.title('BMI by Age')
plt.show()
```

```python
data.bmi.mean()
```

```python
sns.set(style ='whitegrid')

ax = sns.boxplot(x='region', y='bmi',hue='sex', data=data, palette='pastel')

plt.title('BMI by Region and Sex')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='BMI')

plt.show()

```
