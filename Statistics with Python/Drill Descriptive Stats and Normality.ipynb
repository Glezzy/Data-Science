```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
random1 = np.random.poisson(6, 100)

plt.hist(random1)

plt.show()
```

```python
random1.mean()
```

```python
random1.std()
```

```python
plt.hist(random1, bins=20, color='gray')

# plot mean 
plt.axvline(random1.mean(), color='b', linestyle='solid', linewidth=2)

# plot std both ways
plt.axvline(random1.mean() + random1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(random1.mean() - random1.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Using the mean is a good measure of central tendency. The standard
# deviation is not even within 1 std. The standard deviation is not a 
# good measure of the variance.
```

```python
random5 = np.random.standard_cauchy(100)
random5 = random5[(random5>-25) & (random5<25)]

plt.hist(random5)
plt.show()
```

```python
plt.hist(random5, bins=20, color='gray')

# plot mean 
plt.axvline(random5.mean(), color='b', linestyle='solid', linewidth=2)

# plot std both ways
plt.axvline(random5.mean() + random5.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(random5.mean() - random5.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Mean and std are good measures for central tendency and variance.
```

```python
# setting variables
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
# It appears that this dataset has normality since the central tendency
# and variance is well distributed.
```
