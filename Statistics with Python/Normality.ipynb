```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Making a standard normally distributed variable with 1000 observations with a standard devuation of 1 and putting it in a data frame.

mean = 0
sd = 1 
n = 1000

df = pd.DataFrame({'rand': np.random.normal(mean, sd, 1000)})

df.hist()

plt.show
```

```python
# Making 2 variables

rand1 = np.random.normal(50, 300, 1000)
rand2 = np.random.poisson(1, 1000)

# Sorting the valies in ascending order

rand1.sort()
rand2.sort()

# Making a standard normally distributed variable with 1000
# observations, mean of 0, std of 1 as our comparison
norm = np.random.normal(0, 1, 1000)

# Sorting the values in ascending order
norm.sort()
```

```python
plt.plot(norm, rand1, "o")
plt.show()
```

```python
plt.plot(norm, rand2, "o")
plt.show()
```

```python
# Rand1 is normally distributed while Rand2 is not. 
```

```python
plt.hist(rand1, bins=20, color='c')
# vertical line of mean
plt.axvline(rand1.mean(), color='b', linestyle='solid', linewidth=2)
# one std away from mean, above and below
plt.axvline(rand1.mean() + rand1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand1.mean() - rand1.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```
