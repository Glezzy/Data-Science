# KNN Regression vs Multiple Linear Regression



### Question:

   How well do both regression models predict a happiness score                                based on the variables freedom, family, genorosity, and economy.  

### Data:

Data was pulled from kaggle. We will be using the happiness data from 2017 to predict both of our models. 

### Content:

The happiness scores and rankings are from the Gallup World Poll. The scores are found by asking pollers to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0. Then rate there life based on that scale that they have made for themselves. The interesting thing about this data is that each of the factors contribute to making the life evaluations higher in each country than they are in a Dystopia, which in this dataset is a hypotetical country that has values equal to the world's lowest national averages for each of the factors.




### Models:

K Nearest Neighbors Regression

Similar to KNN for classification except that the regression version is used for estimating a continuous variable. KNN Regression can be conducted via the following steps [1].

  1. Compute the Euclidean distance from the query data to the trained data set.
  
  2. Order the trained data set by increasing distance.
  
  3. Find a heuristically optimal number k of nearest neighbors, based on Root Mean Squared Error. This is done using cross validation.
  
  4. Calculate an inverse distance weighted average with the k-nearest multivariate neighbors.

Multiple Regression

######  $$Happiness = \alpha + \beta_1Economy+ \beta_2Family + \beta_3Freedom + \beta_4Generosity$$



Family - Social support (or having someone to count on in times of trouble) is the national average of the binary responses (either 0 or 1) to the Gallup World Poll (GWP) question “If you were in trouble, do you have relatives or friends you can count on to help you whenev-er you need them, or not?”

Freedom - To make life choices is the national average of binary responses to the GWP question “Are you satisfied or dissatisfied with your freedom to choose what you do with your life?”

Generosity - Is the residual of regressing the national average of GWP responses to the question “Have you donated money to a charity in the past month?” on GDP per capita.

Economy= The extent to which GDP contributes to the Happiness score, specifically using Purchasing Power Parity.





```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
import statsmodels.formula.api as smf
%matplotlib inline

sns.set_style('whitegrid')
```

```python
happy = pd.read_csv('world_happiness2017.csv')
happy.head()
```

```python
happy = happy[['Country', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Family','Generosity',
              'Freedom']]
```

```python
happy = happy.rename(columns={'Happiness.Score' : 'happy_score'})
happy = happy.rename(columns={'Economy..GDP.per.Capita.' : 'economy'})
happy = happy.fillna(0)
happy.head()
```

```python
happy.describe()
```

```python
sns.pairplot(happy)
```

```python
import copy

exclude = ['Country']

happy_copy = copy.deepcopy(happy)

corr = happy.loc[:, happy.columns.difference(exclude)].corr()
hmap = sns.heatmap(corr, annot=True)
```

Looks like we need to transform Family, Freedom and Generosity. For Family and Freedom we will square the values, while for Generosity we will take the square root. Hopefully these transformations will give us a more normal distribution.

```python
fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
sm.graphics.qqplot(happy.Family, line='45', fit=True, ax=ax)
ax.set_title('Raw : Family')

ax = fig.add_subplot(2, 2, 2)
sm.graphics.qqplot(np.square(happy.Family), line='45', fit=True, ax=ax)
ax.set_title('Square : Family')
```

```python
sns.distplot(np.square(happy.Family))
plt.title('Distribution of Family Squared Values')
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
sm.graphics.qqplot(happy.Freedom, line='45', fit=True, ax=ax)
ax.set_title('Raw : Freedom')

ax = fig.add_subplot(2, 2, 2)
sm.graphics.qqplot(np.square(happy.Freedom), line='45', fit=True, ax=ax)
ax.set_title('Square : Freedom')
```

```python
sns.distplot(np.square(happy.Freedom))
plt.title('Distribution of Freedom Squared Values')
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
sm.graphics.qqplot(happy.Generosity, line='45', fit=True, ax=ax)
ax.set_title('Raw : Generosity')

ax = fig.add_subplot(2, 2, 2)
sm.graphics.qqplot(np.sqrt(happy.Generosity), line='45', fit=True, ax=ax)
ax.set_title('Square root : Generosity')
```

```python
sns.distplot(np.sqrt(happy.Generosity))
plt.title('Distribution of Generosity Square root Values')
```

```python
happy['Square_Family'] = np.square(happy.Family)

happy['Square_Freedom'] = np.square(happy.Freedom)

happy['Square_root_Generosity'] = np.sqrt(happy.Generosity)

exclude = ['Country', 'Family', 'Freedom', 'Generosity']
```

### New Multiple Regression Formula

New regression formula after transformations.

$$Happiness = \alpha + \beta_1Economy+ \beta_2Family^2 + \beta_3Freedom^2 + \beta_4\sqrt{Generosity}$$

```python

linear_formula = 'happy_score ~ economy+Square_Family+Square_Freedom+Square_root_Generosity'
lm = smf.ols(formula=linear_formula, data=happy).fit()
print(lm.summary())
```

### KNN Regression

```python
happy_copy.head()
```

```python
x = happy_copy[['economy', 'Family', 'Generosity', 'Freedom']]
y = happy_copy['happy_score']
```

```python
from sklearn import neighbors
from sklearn.model_selection import cross_val_score
from sklearn.cross_validation import train_test_split
from sklearn import metrics

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)

# build model
knn = neighbors.KNeighborsRegressor(n_neighbors=10)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)

print('\nR-squared:')
print('%.3f' % (knn.score(x_train, y_train)))

print('\nRMSE: %.3f' % (np.sqrt(metrics.mean_squared_error(y_test, y_pred))))

score = cross_val_score(knn, x, y, cv=5, scoring='neg_mean_squared_error')
print('\nAccuracy: %0.2f (+/- %0.2f)' % (score.mean(), score.std()* 2))
```

### Conclusion

Both of our models had very similar results with the multiple regression model getting slightly better results then the KNN regression. 

A difference between the models that should be noted is that in the multiple regression model the regression line is a straight line. Meanwhile in the KNN regression the line will not be straight because of the windowed average used for K.

Now the interesting part about this is that the KNN regression didn't require transformations of variables and still got very similar results. Also in the multiple linear regression errors need to have a Gaussian distribution. Overall I'd say the KNN regression since it took significantly less steps to perform the regression compared to the multiple linear regression.
