```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
pairs = pd.read_csv('forex.csv', index_col=[0], parse_dates=True)
```

```python
pairs.head(50)
```

```python
ewma = pd.Series.ewm

avg1 = ewma(pairs['EURO_JPY'], span=10).mean()

avg2 = ewma(pairs['AUD_JPY'], span=10).mean()

avg3 = ewma(pairs['EUR_GBP'], span=10).mean()

avg4 = ewma(pairs['GBP_JPY'], span=10).mean()

avg5 = ewma(pairs['EUR_CHF'], span=10).mean()

avg6 = ewma(pairs['CAD_NZD'], span=10).mean()

plt.plot(avg1)
plt.plot(pairs['EURO_JPY'], color='green', alpha=.3)
```
