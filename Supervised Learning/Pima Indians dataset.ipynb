```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
# RNG to reproduce results
np.random.seed(7)
```

```python
diabetes = np.loadtxt("pima_indians_diabetes.csv", delimiter=",")

X = diabetes[:,0:8]
Y = diabetes[:,8]
```

### Define Model

We will be modeling in Keras and creating a Sequential model where we'll add one layer at a time until we are satisfied with our network topology. To do this we first need to make sure the input laer has the right number of inputs. We'll start with 8 inputs and define them using the Dense class.

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

### Compile Model

Once defined we can now compile the model using efficient numerical libraries under the covers of Theanos or Tensorflow. We'll need to add some more properties to train the network. We will use a logorithmic loss for our loss function and also use the default gradient descent algorithm.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
### fitting the model
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# Remove random seed so we can reproduce results
np.random.seed(7)

# load dataset
diabetes = np.loadtxt("pima_indians_diabetes.csv", delimiter=",")

# split into input and output
X = diabetes[:,0:8]
Y = diabetes[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# dataset with automatic validation set
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

np.random.seed(7)

diabetes = np.loadtxt("pima_indians_diabetes.csv", delimiter=",")

X = diabetes[:,0:8]
Y = diabetes[:,8]

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10)
```
