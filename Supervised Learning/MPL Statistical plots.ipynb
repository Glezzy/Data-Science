```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate 1000 random values
x = np.random.normal(10, 5, 1000)
# Plot them as a histogram.
plt.hist(x)
plt.show()
```

```python
# Random data

x = np.random.normal(10, 5, 1000)

# Buildour histogram. Let's go ahead and set the color too!
plt.hist(x, bins=40, color='red')
plt.title('Defualt Bin Placement Demo')
plt.xlabel('Random Values')
plt.show()
```
